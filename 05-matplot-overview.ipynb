## matplot and numpy

```python
import math
```

```python
t = np.arange(0, 2.5, 0.1)
y1 = np.sin(math.pi*t)
y2 = np.sin(math.pi*t + math.pi/2)
y3 = np.sin(math.pi*t - math.pi/2)
```

```python
plt.plot(t, y1, 'b*', t, y2, '^', t, y3, 'ys')
```
